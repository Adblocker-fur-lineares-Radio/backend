```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
raw_data = pd.read_csv("adtime.csv", delimiter=";")
```

```python
data = raw_data.copy()
data['date'] = pd.to_datetime(data['date'])
```

```python
df = data.loc[data['typeflag'] == "start_of_ad"]
sns.barplot(x=["1Live", "WDR2", "Bayern1", "100,5", "RST", "AC", "BremenNext"], y= [df.query('stationName == "1Live"').stationName.count()/30, 
                                                                                    df.query('stationName == "WDR 2"').stationName.count()/30, 
                                                                                    df.query('stationName == "BAYERN 1"').stationName.count()/30,
                                                                                    df.query('stationName == "100,5"').stationName.count()/30,
                                                                                    df.query('stationName == "Radio RST"').stationName.count()/30,
                                                                                    df.query('stationName == "Antenne AC"').stationName.count()/30,
                                                                                    df.query('stationName == "BremenNext"').stationName.count()/30,
                                                                                   ]).set(title='Durschnittliche Anzahl der Werbungen pro Tag')
```

```python
rst = data.loc[data['stationName'] == "Radio RST"]
rst = rst.loc[rst['typeflag'] == "start_of_ad"].drop(["stationName", "typeflag"], axis=1)
arr = rst['date'].dt.minute.value_counts(sort=False).to_frame().sort_values(by=['date']).reset_index()
for x in range(60):
    if x not in arr['date'].values:
        arr.loc[len(arr)] = [x, 0]
arr1 = arr.sort_values(by=['date'])

sns.barplot(x = np.arange(60), y = arr1['count']).set(title='In welchen Minuten traten die Werbungen im letzten Monat bei RadioRST auf?')
fig = plt.gcf()
fig.set_size_inches(20, 5)
plt.show()
 
```

```python
r = arr1.rename(columns={'count': 'RadioRst'})
a = arr2.rename(columns={'count': 'AntenneAC'})
br = arr3.rename(columns={'count': 'BremenNext'})
dashnie = arr4.rename(columns={'count': '100,5'})
ba = arr5.rename(columns={'count': 'Bayern1'})
wd = arr6.rename(columns={'count': 'WDR2'})
li = arr7.rename(columns={'count': '1Live'})

result = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(r,a,on='date',how='outer'),br,on='date',how='outer'),dashnie,on='date',how='outer'),ba,on='date',how='outer'),wd,on='date',how='outer'),li,on='date',how='outer')
result = result.drop(["date"], axis=1)

sns.lineplot(data=result).set(title='In welchen Minuten traten die Werbungen im letzten Monat auf?', xlabel='Minute', ylabel='Anzahl der Werbungen')
fig = plt.gcf()
fig.set_size_inches(20, 5)
plt.show()
```
